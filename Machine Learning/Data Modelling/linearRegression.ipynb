```python
from sklearn.model_selection import train_test_split
```

```python
#so those one you already know them
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#load data
data=pd.read_csv('Stipend_Data.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data['ReserachExperience'],data['Stipend'])
```

```python
x=data['ReserachExperience'].values
y=data['Stipend'].values
```

```python
x.shape
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
Xtrain,Xtest,Ytrain,Ytest= train_test_split(x,y,test_size = 0.2,random_state=2)
```

```python
x.shape
```

```python
#here we are using the model that we import from sklearn
model=Lasso(alpha=0.1,max_iter=1000)
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html#sklearn.linear_model.Lasso
#model = LinearRegression()
history = model.fit(Xtrain,Ytrain)
train_score = history.score(Xtrain,Ytrain)
test_score = history.score(Xtest,Ytest)
```

>
>This is how looks like simple linear regression(with one independante variable) and mutivariate linear regression formulas  
>Linear tries to predict the best suitable intercept and coefficients to make a predict your y value which is the dependant variable

>Where 1 is the slope and o is the y-intercept similar to the equation of a line. The values 1 and o must be chosen so that they minimize the error. To check the error we have to calculate the sum of squared error and tune the parameters to try to reduce the error.

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
# print the coefficients
print(model.intercept_)
print(model.coef_)
```

>score(X, y, sample_weight=None)[source]
>

```python
print('train_score:-{}\ntest_score:-{}'.format(train_score,test_score))
```

```python
pred = model.predict(Xtest)
total_pred = model.predict(x)
plt.plot(Xtest,pred)
plt.scatter(Xtest,Ytest,color = "r")
plt.show()
plt.scatter(x,y)
plt.plot(x,total_pred,color = "r")
plt.title("All_Data")
plt.show()
```

### How to evaluate Linear regression
>The sklearn. metrics module implements several loss, score, and utility functions to measure classification performance.

```python

from sklearn import metrics
y_pred=model.predict(Xtest)
# RMSE
print(np.sqrt(metrics.mean_squared_error(Ytest, y_pred)))
```
